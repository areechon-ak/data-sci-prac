```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Creating list to append tweet data to
tweets_list = []

max = 20000

# Using TwitterSearchScraper to scrape data and append tweets to list
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('‡πÑ‡∏õ‡∏£‡∏©‡∏ì‡∏µ‡∏¢‡πå‡πÑ‡∏ó‡∏¢ since:2022-01-01 until:2022-01-31').get_items()):
    if i>max:
        break
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, \
                        tweet.url, tweet.retweetCount, tweet.likeCount, tweet.quoteCount, \
                        tweet.retweetedTweet, tweet.quotedTweet, tweet.media])
```

```python
# Creating a dataframe from the tweets list above
df2022 = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'URL', 'retweetCount', 'likeCount', 'quoteCount', 'retweetedTweet', 'quotedTweet', 'media'])
```
